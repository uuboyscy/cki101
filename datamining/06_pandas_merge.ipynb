```python
import pandas as pd
```

```python
path_staff = "staff.csv"
path_staff_detail = "staff_detail.csv"
path_staff_detail_different_col = "staff_detail_different_col.csv"
path_staff_samecol = "staff_samecol.csv"

df_staff = pd.read_csv(path_staff)
df_staff_detail = pd.read_csv(path_staff_detail)
df_staff_detail_different_col = pd.read_csv(path_staff_detail_different_col)
df_staff_samecol = pd.read_csv(path_staff_samecol)
```

```python
df_staff_detail.merge(
    df_staff,
    how="left",
    left_on="ID",
    right_on="ID",
)
```

```python
df_staff_detail_different_col.merge(
    df_staff_samecol,
    how="left",
    left_on="EmpNum",
    right_on="ID",
    suffixes=["_left", "_right"]
)[
    [
        "EmpNum",
        "Level",
        "Name",
        "Salary",
    ]
]
```
