```python
"""
select
    portno,
    eff_dt,
    sum(percent_tna)
from df
group by portno, eff_dt
order by portno, eff_dt desc
"""
df_sorted = (
    df
    .groupby(["portno", "eff_dt"], as_index=False)["percent_tna"]
    .sum()
    .sort_values(["portno", "eff_dt"], ascending=False)
)

df_sorted
```
